### Поставленные лайки

```python
query = '''
           select u.user_id,
                  u.name, 
                  count(l.user_id) as "поставленные лайки"
            from users u
            left join likes l on u.user_id = l.user_id
            group by u.user_id, u.name;
            
            '''
pd.io.sql.read_sql(query, con = engine) 
```

### Полученные лайки

```python
query = '''
           select u.user_id,
                  u.name, 
                  count(l.user_id_give) as "полученные лайки"
            from users u
            left join likes l on u.user_id = l.user_id_give
            group by u.user_id, u.name;
            
            '''
pd.io.sql.read_sql(query, con = engine) 
```

### Взаимные лайки

```python
query = '''
           select u.user_id,
                  u.name, 
                  count(l.user_id_give) as "взаимные лайки"
            from users u
            left join likes l on u.user_id = l.user_id_give
            where u.user_id = l.user_id_give
            group by u.user_id, u.name;
            
            '''
pd.io.sql.read_sql(query, con = engine) 
```

### Самые популярные пользователи (те, у кого больше всего лайков)

```python
query = '''
           select u.user_id,
                  u.name, 
                  count(l.user_id_give) as "полученные лайки"
            from users u
            left join likes l on u.user_id = l.user_id_give
            group by u.user_id, u.name
            order by "полученные лайки" desc
            limit 5;
            
            '''
pd.io.sql.read_sql(query, con = engine) 
```
